```python
import pandas as pd
import numpy as np
```

```python
fighters = pd.read_csv('../CSV_Files/Fighters.csv')
fighters
```

```python
ChimaevFight = pd.read_csv('../CSV_Files/ChimaevFights.csv')
ChimaevFight
```

```python
StricklandFight = pd.read_csv('../CSV_Files/StricklandFights.csv')
StricklandFight
```

```python
ChimaevDS = fighters[fighters['Name'] == 'Khamzat Chimaev']
StricklandDS = fighters[fighters['Name'] == 'Sean Strickland']

for i in range(1,14):
    ChimaevDS.insert(i+13, ChimaevFight.columns[i], 0)
    StricklandDS.insert(i+13, StricklandFight.columns[i], 0)
```

```python
columns_names = list(fighters.columns) + [col for col in ChimaevFight.columns if col != 'Fight']

for i in range(0, 9):
    data = list(fighters.iloc[i+1, :]) + list(ChimaevFight.iloc[i, 1::])
    dataframe = pd.DataFrame(data, index=columns_names).transpose()
    ChimaevDS = pd.concat([ChimaevDS, dataframe])

ChimaevDS = ChimaevDS.reset_index(drop=True)
ChimaevDS
```

```python
data = list(fighters.iloc[11, :]) + list(StricklandFight.iloc[0, 1::])
df = pd.DataFrame(data, index=columns_names).transpose()
StricklandDS = pd.concat([StricklandDS, df])

data = list(fighters.iloc[1, :]) + list(StricklandFight.iloc[1, 1::])
df = pd.DataFrame(data, index=columns_names).transpose()
StricklandDS = pd.concat([StricklandDS, df])

data = list(fighters.iloc[12, :]) + list(StricklandFight.iloc[2, 1::])
df = pd.DataFrame(data, index=columns_names).transpose()
StricklandDS = pd.concat([StricklandDS, df])

data = list(fighters.iloc[1, :]) + list(StricklandFight.iloc[3, 1::])
df = pd.DataFrame(data, index=columns_names).transpose()
StricklandDS = pd.concat([StricklandDS, df])

for i in range(4, 17):
    data = list(fighters.iloc[i+9, :]) + list(StricklandFight.iloc[i, 1::])
    df = pd.DataFrame(data, index=columns_names).transpose()
    StricklandDS = pd.concat([StricklandDS, df])

data = list(fighters.iloc[3, :]) + list(StricklandFight.iloc[17, 1::])
df = pd.DataFrame(data, index=columns_names).transpose()
StricklandDS = pd.concat([StricklandDS, df])

for i in range(18, 24):
    data = list(fighters.iloc[i+8, :]) + list(StricklandFight.iloc[i, 1::])
    df = pd.DataFrame(data, index=columns_names).transpose()
    StricklandDS = pd.concat([StricklandDS, df])

StricklandDS = StricklandDS.reset_index(drop=True)
StricklandDS
```
