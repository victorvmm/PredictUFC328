```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, f1_score

%run DataCleaning.ipynb
```

Criação do modelo

```python
model = LogisticRegression(penalty='l1', solver='liblinear', tol=1e-3, random_state=42) # perda = l1 e solver = liblinear, adequado para modelos de dataset pequeno
```

Divisão em conjuntos de treinamento e de teste para validar modelo

```python
train_data = StricklandDS.iloc[0:19, 1:-1].values # últimas 19 lutas como conjunto de treinamento
train_target = StricklandDS.iloc[0:19, -1].values

test_data = StricklandDS.iloc[19:, 1:-1].values # primeiras 6 lutas como conjunto de testes 
test_target = StricklandDS.iloc[19:, -1].values
```

```python
model.fit(train_data, train_target) # Treinamento do modelo

pred = model.predict(test_data) # Predições do conjunto de teste
```

```python
print(f'Precisão do modelo: {accuracy_score(test_target, pred)*100:.0f}%') # Porcentagem de predições positivas (vitórias)
print(f'Recall do modelo: {recall_score(test_target, pred)*100:.0f}%') # Porcentagem de 
print(f'F1-score do modelo: {f1_score(test_target, pred)*100:.0f}%')
```
