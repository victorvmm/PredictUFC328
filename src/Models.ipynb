```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, recall_score, f1_score

%run DataCleaning.ipynb
```

Model creation

```python
model = LogisticRegression(penalty='l1', solver='liblinear', tol=1e-3, random_state=42) 
```

Train and test split to validate model

```python
train_data = StricklandDS.iloc[0:19, 1:-1].values # Last 19 fights as train set
train_target = StricklandDS.iloc[0:19, -1].values

test_data = StricklandDS.iloc[19:, 1:-1].values # First 6 fights as 
test_target = StricklandDS.iloc[19:, -1].values
```

```python
model.fit(train_data, train_target) # Model training

pred = model.predict(test_data) # Predicting using test set
```

```python
print(f'Model precision: {accuracy_score(test_target, pred)*100:.0f}%') 
print(f'Model recall: {recall_score(test_target, pred)*100:.0f}%')
print(f'Model F1-Score: {f1_score(test_target, pred)*100:.0f}%')
```
